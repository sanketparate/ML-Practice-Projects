```python
X=salary_data.drop("salary",axis=1)
y=salary_data[["salary"]]
```

```python
# Lets bring the dataset features into same scle
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X= scaler.fit_transform(X)
```

<b>Splitting the data into Train and Test</b>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.30)
```

```python
# We are using R2 score to evaluate
from sklearn.metrics import r2_score
def max_r2_score(re,X,y):
    max_r2_score=0
    for r_state in range(42,100):
        X_train,X_test, y_train, y_test= train_test_split(X,y,test_size=0.30, random_state=r_state)
        re.fit(X_train,y_train)
        y_pred= re.predict(X_test)
        score=r2_score(y_test,y_pred)
        print("R2_Score corresponding to",r_state," is",score)
        if score>max_r2_score:
            max_r2_score=score
            final_r2_score=r_state
    print("Maximum R2_Score corresponding:",final_r2_score," and it is :",round((max_r2_score),3))
```

# Machine Learning Models:

We are using Linear Regression, Decision tree Regressor, KNN Regressor, Random Forest Regressor and Gradient Boosting Regressor.

<b>Linear Regression</b>

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
max_r2_score(lr,X,y)
```

```python
predlr= lr.predict(X_test)
model1= lr.score(X_test, y_test)
print("The accuracy of the Linear Regression Model is:",round((model1),3))
```

```python
from sklearn import metrics
lr1=np.sqrt(metrics.mean_squared_error(y_test,predlr))
lr1
```

<b>Decision Tree Regressor</b>

```python
from sklearn.tree import DecisionTreeRegressor
drt=DecisionTreeRegressor()
max_r2_score(drt,X,y)
```

```python
preddrt= drt.predict(X_test)
model2= drt.score(X_test, y_test)
print("The accuracy of the Decision Tree Regressor Model is:",round((model2),3))
```

```python
from sklearn import metrics
drt1=np.sqrt(metrics.mean_squared_error(y_test,preddrt))
drt1
```

<b>KNN Regressor</b>

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
parameters= {"n_neighbors":range(2,10)}
knn= KNeighborsRegressor()
knn1= GridSearchCV(knn, parameters, cv=5)
knn1.fit(X,y)
knn1.best_params_
```

```python
knn1= KNeighborsRegressor(n_neighbors=5)
max_r2_score(knn1, X,y)
```

```python
predknn= knn1.predict(X_test)
model3= knn1.score(X_test, y_test)
print("The accuracy of the KNN Regressor Model is:",round((model3),3))
```

```python
knn_1=np.sqrt(metrics.mean_squared_error(y_test,predknn))
knn_1
```

<b>Random Forest Regressor</b>

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,random_state=0)
max_r2_score(rf,X,y)
```

```python
predrf= rf.predict(X_test)
model4= rf.score(X_test, y_test)
print("The accuracy of the Random Forest Regressor Model is:",round((model4),3))
```

```python
rf_1=np.sqrt(metrics.mean_squared_error(y_test,predrf))
rf_1
```

<b>Gradient Boosting Regressor</b>

```python
from sklearn.ensemble import GradientBoostingRegressor
gbcl = GradientBoostingRegressor(n_estimators = 50, learning_rate = 0.09, max_depth=5)
max_r2_score(gbcl,X,y)
```

```python
predgbcl= gbcl.predict(X_test)
model5=gbcl.score(X_test, y_test)
print("The accuracy of the Gradient Boosting Regressor Model is:",round((model5),3))
```

```python
gbcl_1=np.sqrt(metrics.mean_squared_error(y_test,predgbcl))
gbcl_1
```
